```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('asset/resized_livephotoes/4.jpg')
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
model = tf.keras.models.load_model('asset/models/occ_or_mpty_model.h5')
# Verific dac modelul a fost ncrcat corect
if model is None:
    print("Modelul nu a putut fi ncrcat.")
    exit()
```

```python
resize = tf.image.resize(img, (256,256))
prediction = model.predict(np.expand_dims(resize/255, axis=0))
```

```python
print("Rezultatul prediciei:", prediction)

if prediction > 0.5: 
    print(f'Predicted class is occ')
    
else:
    print(f'Predicted class is mpty')
```
