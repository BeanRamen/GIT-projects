```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('asset/resized_livephotoes/resized_5.jpg')
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
model = tf.keras.models.load_model('asset/models/occ_or_mptyy.h5')
# Verifică dacă modelul a fost încărcat corect
if model is None:
    print("Modelul nu a putut fi încărcat.")
    exit()
```

```python
prediction = model.predict(np.expand_dims(resize, axis=0))
```

```python
print("Rezultatul predicției:", prediction)

if prediction > 0.5: 
    print(f'Predicted class is occ')
    
else:
    print(f'Predicted class is mpty')
```
